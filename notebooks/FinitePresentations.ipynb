```python
import networkx
def left_cayley(S):
    G=networkx.DiGraph()
    for i in range(S.size()):
        for j in range(S.nr_generators()):
            G.add_edge(str(i),str(S.left(i,j)))
    return G
```

```python
G=left_cayley(S)
```

```python
networkx.draw(G)
```

```python
G.nodes()
```

```python
S.size()
```

```python
def right_cayley(S):
    G=networkx.DiGraph()
    for i in range(S.size()):
        for j in range(S.nr_generators()):
            G.add_edge(str(i),str(S.right(i,j)))
    return G
```

```python
G=right_cayley(S)
```

```python
networkx.draw(G)
```

```python
G.nodes()
```
